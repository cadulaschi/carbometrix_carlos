## Data analysis if the biggest automotive companies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Carbometrix_Project/automotive_data.csv')
data = data.transpose()
```

```python
data
```

```python
emissions = data.iloc[4:]
emissions = emissions.replace('<Not Applicable>', np.nan)
name_emissions = list(emissions.index)
emissions
```

```python
i = 0
mean_emissions = np.zeros(len(emissions))
for s in np.array(emissions, dtype=float):
    s = s[np.logical_not(np.isnan(s))] #do not count companies with nan 
    mean_emissions[i] = np.mean(s)        
    i += 1
```

```python
y_pos = np.arange(len(name_emissions))

fig, ax = plt.subplots(figsize=(17,10))

ax.barh(y_pos, mean_emissions)
ax.set_yticks(y_pos)
ax.set_yticklabels(name_emissions)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('CO2eq')
ax.set_title('Mean of Emissions (all scopes, absolute value)')

plt.show()
```
